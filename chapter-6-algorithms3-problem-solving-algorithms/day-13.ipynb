## Unit 30: Greedy Approach
## Greedy Approach

Greedy algorithms make the best choice at each step with the hope of reaching a global optimum

### Key Idea:
- Local optimal choice
- No backtracking
- Fast and simple

### Steps:
1. Selection
2. Feasibility Check
3. Solution Check

### Applications:
- Coin Change Problem
- Activity Selection Problem

### Coin Change Problem

Goal: Minimize number of coins used to form a given amount

```python
def coin_change(coins, amount):
    result = []
    i = 0

    while amount > 0:
        if coins[i] <= amount:
            amount -= coins[i]
            result.append(coins[i])
        else:
            i += 1

    return result


# user input
coins = [500, 100, 50, 10]
amount = int(input("Enter amount: "))

ans = coin_change(coins, amount)

print("Coins used:", ans)
print("Total coins:", len(ans))
```

### Activity Selection Problem

Goal: Select maximum number of non-overlapping activities

```python
def activity_selection(start, finish):
    n = len(start)
    selected = [0]

    i = 0
    for j in range(1, n):
        if start[j] >= finish[i]:
            selected.append(j)
            i = j

    return selected


# input
start = list(map(int, input("Enter start times: ").split()))
finish = list(map(int, input("Enter finish times: ").split()))

result = activity_selection(start, finish)

print("Selected activities:", result)
print("Max meetings:", len(result))
```

## Unit 31: Divide and Conquer

Divide and Conquer solves problems by breaking them into smaller subproblems.

### Steps:
1. Divide
2. Conquer
3. Combine

### Examples:
- Binary Search
- Merge Sort
- Quick Sort

```python
def binary_search(nums, low, high, x):
    if low > high:
        return -1

    mid = (low + high) // 2

    if nums[mid] == x:
        return mid
    elif x < nums[mid]:
        return binary_search(nums, low, mid - 1, x)
    else:
        return binary_search(nums, mid + 1, high, x)


nums = list(map(int, input("Enter sorted list: ").split()))
x = int(input("Enter element to search: "))

res = binary_search(nums, 0, len(nums)-1, x)

print("Index:", res)
```

### Minimum Path Sum (Triangle)

Goal: Find the minimum path sum from top to bottom  
Each step moves to adjacent elements in the next row

```python
def find_min(triangle, row, col):
    if row == len(triangle):
        return 0

    return triangle[row][col] + min(
        find_min(triangle, row+1, col),
        find_min(triangle, row+1, col+1)
    )


# input
n = int(input("Enter rows: "))
triangle = []

for i in range(n):
    row = list(map(int, input().split()))
    triangle.append(row)

print("Minimum path sum:", find_min(triangle, 0, 0))
```
