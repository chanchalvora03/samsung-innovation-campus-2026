## Unit 32: Dynamic Programming

Dynamic Programming stores results of subproblems to avoid recomputation.

### Key Concepts:
- Overlapping subproblems
- Memoization
- Bottom-up approach

### Examples:
- Fibonacci
- Pascal’s Triangle

### Fibonacci using Dynamic Programming

Goal: Compute nth Fibonacci number efficiently

```python
def fib(n):
    dp = [0]*(n+1)
    dp[1] = 1

    for i in range(2, n+1):
        dp[i] = dp[i-1] + dp[i-2]

    return dp[n]


n = int(input("Enter n: "))
print("Fibonacci:", fib(n))
```

This avoids exponential recursion and reduces time complexity to O(n)

### Pascal’s Triangle

Each value is the sum of two values above it

```python
def pascal(n):
    B = [[0]*(i+1) for i in range(n+1)]

    for i in range(n+1):
        for j in range(i+1):
            if j == 0 or j == i:
                B[i][j] = 1
            else:
                B[i][j] = B[i-1][j-1] + B[i-1][j]

    return B


n = int(input("Enter n: "))
triangle = pascal(n)

for row in triangle:
    print(row)
```

This is a bottom-up dynamic programming approach

## Unit 33: Backtracking

Backtracking builds solution step-by-step and removes invalid ones.

### Key Concepts:
- Depth First Search
- Pruning
- Promising

### N-Queens Problem

Goal: Place N queens on a chessboard so that no two attack each other

```python
def is_safe(board, row, col):
    for i in range(row):
        if board[i] == col or abs(board[i] - col) == abs(i - row):
            return False
    return True


def solve(board, row, n):
    if row == n:
        print(board)
        return

    for col in range(n):
        if is_safe(board, row, col):
            board[row] = col
            solve(board, row+1, n)


n = int(input("Enter N: "))
board = [-1]*n

solve(board, 0, n)
```

Backtracking tries all configurations and prunes invalid ones early
