## What is Sorting?
Sorting arranges data in ascending or descending order for easier searching and processing.

```
numbers=[64,34,25,12,22,11,90]
print("Original:",numbers)
```

## Bubble Sort

```
arr=[64,34,25,12,22,11,90]
n=len(arr)
for i in range(n):
    for j in range(0,n-i-1):
        if arr[j]>arr[j+1]:
            arr[j],arr[j+1]=arr[j+1],arr[j]
print("Sorted:",arr)
```

## Selection Sort

```
arr=[64,25,12,22,11]
for i in range(len(arr)):
    min_idx=i
    for j in range(i+1,len(arr)):
        if arr[j]<arr[min_idx]:
            min_idx=j
    arr[i],arr[min_idx]=arr[min_idx],arr[i]
print(arr)
```

## Insertion Sort

```
arr=[12,11,13,5,6]
for i in range(1,len(arr)):
    key=arr[i]
    j=i-1
    while j>=0 and key<arr[j]:
        arr[j+1]=arr[j]
        j-=1
    arr[j+1]=key
print(arr)
```

## Time Complexity
| Algorithm | Best | Average | Worst |
|---|---|---|---|
|Bubble|O(n)|O(n²)|O(n²)|
|Selection|O(n²)|O(n²)|O(n²)|
|Insertion|O(n)|O(n²)|O(n²)|
