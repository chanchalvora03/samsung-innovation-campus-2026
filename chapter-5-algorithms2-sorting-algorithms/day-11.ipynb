## Bubble Sort, Selection Sort, Insertion Sort

Sorting is the process of arranging data in a specific order (ascending or descending).

## Bubble Sort

- Compare adjacent elements
- Swap if wrong order
- Largest element moves to the end each pass

Time Complexity: O(n²)

```python
def bubble_sort(S):
    n = len(S)
    for i in range(n):
        for j in range(0, n-i-1):
            if S[j] > S[j+1]:
                S[j], S[j+1] = S[j+1], S[j]
        print(f"After pass {i+1}: {S}")
    return S

arr = list(map(int, input("Enter numbers: ").split()))
print("Sorted:", bubble_sort(arr))
```

After each pass, the largest element moves to the end

## Selection Sort

- Find smallest element
- Swap with first unsorted position
- Repeat for remaining list

```python
def selection_sort(S):
    n = len(S)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if S[j] < S[min_idx]:
                min_idx = j
        S[i], S[min_idx] = S[min_idx], S[i]
        print(f"After pass {i+1}: {S}")
    return S

arr = list(map(int, input("Enter numbers: ").split()))
print("Sorted:", selection_sort(arr))
```

It reduces the unsorted part step by step from the beginning

## Insertion Sort

- Builds sorted list step by step
- Inserts element at correct position

```python
def insertion_sort(S):
    for i in range(1, len(S)):
        key = S[i]
        j = i - 1

        while j >= 0 and S[j] > key:
            S[j+1] = S[j]
            j -= 1

        S[j+1] = key
        print(f"After pass {i}: {S}")

    return S

arr = list(map(int, input("Enter numbers: ").split()))
print("Sorted:", insertion_sort(arr))
```

Best Case: O(n)  
Worst Case: O(n²)

### Anagram Problem

Two words are anagrams if they contain same letters in different order.

```python
def is_anagram(a, b):
    return sorted(a) == sorted(b)

a = input("Enter first word: ")
b = input("Enter second word: ")

if is_anagram(a, b):
    print("Anagram")
else:
    print("Not Anagram")
```
