## Merge Sort and Quick Sort

### Merge Sort

- Divide and conquer algorithm
- Split list into halves
- Sort each half recursively
- Merge sorted halves
- Time complexity: O(n log n)

```python
def merge(L, R):
    result = []
    i = j = 0

    while i < len(L) and j < len(R):
        if L[i] < R[j]:
            result.append(L[i])
            i += 1
        else:
            result.append(R[j])
            j += 1

    result.extend(L[i:])
    result.extend(R[j:])
    return result


def merge_sort(S):
    if len(S) <= 1:
        return S

    mid = len(S) // 2
    L = merge_sort(S[:mid])
    R = merge_sort(S[mid:])

    return merge(L, R)


arr = list(map(int, input("Enter numbers: ").split()))
print("Sorted:", merge_sort(arr))
```

### Quick Sort

- Divide and conquer
- Choose pivot
- Partition into:
    smaller than pivot
    larger than pivot
- Recursively sort
- Average: O(n log n)
- Worst: O(n²)

```python
def quick_sort(S):
    if len(S) <= 1:
        return S

    pivot = S[0]
    left = [x for x in S[1:] if x <= pivot]
    right = [x for x in S[1:] if x > pivot]

    return quick_sort(left) + [pivot] + quick_sort(right)


arr = list(map(int, input("Enter numbers: ").split()))
print("Sorted:", quick_sort(arr))
```

## Merge Sort vs Quick Sort

Merge Sort:
- Always O(n log n)
- Uses extra memory

Quick Sort:
- Faster in practice
- In-place sorting
- Worst case O(n²)

## Kth Largest Element

### Q1.
Write an algorithm that finds the Kth largest element in a list of N unordered elements.

### Approach:
1. Sort the list and return the Kth largest element  
2. Use partition (Quick Select idea) to find it efficiently  

Compare which method is more efficient.

### Method 1: Using Sorting
- Sort the list
- Return element at index -k
- Time Complexity: O(n log n)

```python
def kth_largest_sort(arr, k):
    arr.sort()
    return arr[-k]

# User input
arr = list(map(int, input("Enter numbers: ").split()))
k = int(input("Enter k: "))

result = kth_largest_sort(arr, k)
print(f"{k}th largest element is:", result)
```

### Method 2: Using Partition (Quick Select)

- Similar to Quick Sort
- Use pivot and partition
- Only search one side
- Average Time Complexity: O(n)

```python
def partition(arr, low, high):
    pivot = arr[high]
    i = low

    for j in range(low, high):
        if arr[j] >= pivot:   # for kth largest
            arr[i], arr[j] = arr[j], arr[i]
            i += 1

    arr[i], arr[high] = arr[high], arr[i]
    return i


def quick_select(arr, low, high, k):
    if low <= high:
        pivot_index = partition(arr, low, high)

        if pivot_index == k:
            return arr[pivot_index]
        elif pivot_index > k:
            return quick_select(arr, low, pivot_index - 1, k)
        else:
            return quick_select(arr, pivot_index + 1, high, k)


# User input
arr = list(map(int, input("Enter numbers: ").split()))
k = int(input("Enter k: "))

result = quick_select(arr, 0, len(arr)-1, k-1)
print(f"{k}th largest element is:", result)
```

## What I learned

- Divide-and-conquer improves efficiency
- Merge sort is stable but uses extra space
- Quick sort is faster but not always safe
- Quick select is useful for selection problems
