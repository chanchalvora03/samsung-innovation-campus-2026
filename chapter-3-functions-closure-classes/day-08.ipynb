## Closure and Scope

```python
# SCOPE (LEGB RULE)
x = 10

def func():
    x = 20
    print("Local:", x)

func()
print("Global:", x)
```

Shows local vs global scope  
Local variable inside function does not affect global variable

```python
#Global keyword
x = 10

def change():
    global x
    x = 50

change()
print(x)
```

Using global keyword modifies the global variable from inside the function

```python
#NESTED FUNCTION
def outer():
    x = 10
    
    def inner():
        print(x)
    
    inner()

outer()
```

Inner function can access variables from outer function scope

```python
#CLOSURE
def outer():
    x = 10
    
    def inner():
        print(x)
    
    inner()

outer()
```

Closure = inner function remembers variables from outer function even after outer function has finished execution

```python
#CLOSURE WITH PARAMETER
def outer():
    x = 10
    
    def inner():
        return x
    
    return inner

f = outer()
print(f())
```

Returned function retains access to enclosed variable x

```python
city_pop = {
    "A": 9765,
    "B": 3441,
    "C": 2954,
    "D": 1531
}

def analyze(city_pop):
    values = city_pop.values()
    
    print("Max:", max(values))
    print("Min:", min(values))
    print("Diff:", max(values) - min(values))
    print("Avg:", sum(values)/len(values))

analyze(city_pop)
```
