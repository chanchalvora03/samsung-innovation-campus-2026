## Stack

Stack is an Abstract Data Type (ADT) that follows:
LIFO (Last In First Out)

Operations:
- push → add element
- pop → remove element

```python
# Using a list to create stack
def is_empty(stack):
    return len(stack) == 0

def push(stack, item):
    stack.append(item)

def pop(stack):
    if is_empty(stack):
        return None
    return stack.pop()

def peek(stack):
    if is_empty(stack):
        return None
    return stack[-1]

stack = []
push(stack, 'A')
push(stack, 'B')
push(stack, 'C')

print(peek(stack))

print(pop(stack))
print(pop(stack))
print(pop(stack))
print(pop(stack))   # empty now
```

```python
# Using a class to create stack
class Stack:
    def __init__(self):
        self.stack = []

    def is_empty(self):
        return len(self.stack) == 0

    def push(self, item):
        self.stack.append(item)

    def pop(self):
        if self.is_empty():
            return None
        return self.stack.pop()

    def peek(self):
        if self.is_empty():
            return None
        return self.stack[-1]

s = Stack()

s.push(1)
s.push(2)
s.push(3)

print(s.pop())
print(s.peek())
print(s.is_empty())

print(s.pop())
print(s.pop())
print(s.pop())   # empty now
```

## Balanced Parentheses using Stack

A string is balanced if:
1. Opening and closing brackets match
2. Order is correct
3. Stack is empty at the end

```python
def check_parentheses(expr):
    stack = []
    opening = "([{"
    closing = ")]}"

    for char in expr:
        if char in opening:
            stack.append(char)
        elif char in closing:
            if not stack:
                return False
            top = stack.pop()
            if opening.index(top) != closing.index(char):
                return False

    return len(stack) == 0

expr = input("Enter an expression: ")

if check_parentheses(expr):
    print("Balanced")
else:
    print("Not Balanced")
```

Explanation:

- "[]{}()" → correct matching → True  
- "([)]" → wrong order → False  
- "((()))" → properly nested → True  
- "(()" → missing closing bracket → False  

Stack ensures LIFO matching of brackets.

## Queue

A Queue is a linear data structure that follows the FIFO (First In First Out) principle.

The element inserted first is the first one to be removed.

Basic Operations:
- Enqueue → Add an element to the rear (end)
- Dequeue → Remove an element from the front (beginning)

In Python, a queue can be implemented using lists or collections like deque.

```python
class Queue:
    def __init__(self):
        self.queue = []

    def is_empty(self):
        return len(self.queue) == 0

    def enqueue(self, item):
        self.queue.append(item)

    def dequeue(self):
        if self.is_empty():
            return None
        return self.queue.pop(0)

    def size(self):
        return len(self.queue)

q = Queue()

n = int(input("Enter number of elements to enqueue: "))

for i in range(n):
    x = int(input("Enter element: "))
    q.enqueue(x)

print("Queue:", q.queue)

print("Dequeued:", q.dequeue())
print("Queue after dequeue:", q.queue)
```

### Josephus Problem

Given:
- n people standing in a circle
- every k-th person is eliminated

Using a Queue, we simulate the process by rotating elements.

The last remaining person is the survivor.

```python
def josephus(n, k):
    q = Queue()
    result = []

    for i in range(1, n + 1):
        q.enqueue(i)

    count = 1

    while q.size() > 1:
        person = q.dequeue()

        if count % k == 0:
            result.append(person)
        else:
            q.enqueue(person)

        count += 1

    result.append(q.dequeue())
    return result

n = int(input("Enter number of people: "))
k = int(input("Enter step value: "))

print("Elimination order:", josephus(n, k))
```

Each k-th person is removed using queue rotation.

Remaining people are re-added to queue.
Last remaining person is the survivor.

## Sequential Search

Search elements one by one.

Time Complexity:
- Best Case: O(1)
- Worst Case: O(n)

```python
def seq_search(nums, x):
    for i in range(len(nums)):
        if nums[i] == x:
            return i
    return -1

nums = list(map(int, input("Enter numbers (space separated): ").split()))
x = int(input("Enter number to search: "))

result = seq_search(nums, x)

if result != -1:
    print("Element found at index:", result)
else:
    print("Element not found")
```

```python
# Find index of maximum element in a list
def find_max(nums):
    max_index = 0

    for i in range(1, len(nums)):
        if nums[i] > nums[max_index]:
            max_index = i

    return max_index

nums = list(map(int, input("Enter numbers: ").split()))

index = find_max(nums)

print("Max element:", nums[index])
print("Index:", index)
```

### Egg Dropping Problem

Given:
- k eggs
- n floors

Find minimum attempts needed in worst case to determine
the highest floor from which egg does not break.

This is a classic dynamic programming problem.

```python
def egg_drop_2_eggs(n):
    attempts = 0
    step = 1

    while n > 0:
        attempts += 1
        n -= step
        step += 1

    return attempts

floors = int(input("Enter number of floors: "))

print("Minimum attempts (2 eggs):", egg_drop_2_eggs(floors))
```

### Word Count Problem

Count how many times a specific word appears in a sentence.

Steps:
- Take sentence input
- Split into words
- Count occurrences of given word

```python
def word_count(S, x):
    count = 0
    for word in S:
        if word == x:
            count += 1
    return count

S = input("Enter a sentence: ").lower().split()
x = input("Enter word to search: ").lower()

count = word_count(S, x)

print(f"In S, '{x}' appeared {count} times.")
```
