## What is an Algorithm?
An algorithm is a step-by-step procedure to solve a problem.

```
numbers=[10,25,8,42,19]
largest=numbers[0]
for n in numbers:
    if n>largest:
        largest=n
print('Largest:',largest)
```

## Linear Search

```
nums=[4,7,9,12,18,21]
target=12
for i,v in enumerate(nums):
    if v==target:
        print('Found at',i)
        break
```

## Binary Search

```
nums=[2,4,6,8,10,12,14,16]
target=10
low,high=0,len(nums)-1
while low<=high:
    mid=(low+high)//2
    if nums[mid]==target:
        print('Found at',mid)
        break
    elif nums[mid]<target:
        low=mid+1
    else:
        high=mid-1
```
