## Stack (LIFO)

```
stack=[]
stack.append(10)
stack.append(20)
stack.append(30)
print(stack)
print('Popped:',stack.pop())
print(stack)
```

## Queue (FIFO)

```
from collections import deque
q=deque()
q.append('A')
q.append('B')
q.append('C')
print(q)
print('Dequeued:',q.popleft())
print(q)
```

## Linked List (Simple Node)

```
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None

n1=Node(10)
n2=Node(20)
n3=Node(30)
n1.next=n2
n2.next=n3
cur=n1
while cur:
    print(cur.data)
    cur=cur.next
```

## Practice

```
stack=[]
for i in range(5):
    stack.append(i)
while stack:
    print(stack.pop())
```
