## Binary Search

Binary Search works on a sorted list.

It repeatedly divides the search space into half.

Steps:
1. Find middle element
2. Compare with target
3. Move left or right accordingly

Time Complexity:
- Best Case: O(1)
- Worst Case: O(log n)

```python
def binary_search(nums, x):
    low = 0
    high = len(nums) - 1

    while low <= high:
        mid = (low + high) // 2

        if nums[mid] == x:
            return mid
        elif nums[mid] < x:
            low = mid + 1
        else:
            high = mid - 1

    return -1

nums = list(map(int, input("Enter sorted numbers: ").split()))
x = int(input("Enter element to search: "))

result = binary_search(nums, x)

if result != -1:
    print("Element found at index:", result)
else:
    print("Element not found")
```

Note:
Binary search works only on sorted lists.

### Search Insert Position

Given a sorted list and a number x,
find the position where x should be inserted
so that the list remains sorted.

```python
def search_insert_position(nums, x):
    low = 0
    high = len(nums)

    while low < high:
        mid = (low + high) // 2

        if nums[mid] < x:
            low = mid + 1
        else:
            high = mid

    return low

nums = list(map(int, input("Enter sorted numbers: ").split()))
x = int(input("Enter number to insert: "))

pos = search_insert_position(nums, x)

print(f"{x} should be inserted at position {pos}")

nums.insert(pos, x)
print("Updated list:", nums)
```

### Egg Dropping Problem

Find minimum attempts needed to determine
the highest floor from which an egg will not break.

```python
def egg_drop_2_eggs(n):
    attempts = 0
    step = 1

    while n > 0:
        attempts += 1
        n -= step
        step += 1

    return attempts

floors = int(input("Enter number of floors: "))
print("Minimum attempts:", egg_drop_2_eggs(floors))
```

## Summary

- Binary search reduces search space efficiently
- Insert position ensures sorted order
- Egg dropping demonstrates problem-solving logic

# Hash Table

A Hash Table stores data in key-value pairs.

It allows fast insertion, deletion, and lookup.

In Python, dictionaries are used to implement hash tables.

Time Complexity (average case):
- Insert: O(1)
- Search: O(1)
- Delete: O(1)

```python
data = {
    "name": "Alicia",
    "age": 20,
    "course": "Python"
}

print(data["name"])

data["age"] = 21
data["city"] = "Blr"

print(data)

del data["course"]
print(data)
```

### Example of Hash Table

The key value for each book is calculated using ASCII values.

Each character is converted using ord() and summed to form the key.

```python
books = [
    "The Little Prince",
    "The Old Man and the Sea",
    "The Little Mermaid",
    "Beauty and the Beast",
    "The Last Leaf"
]

for book in books:
    key = sum(map(ord, book))
    print(key, book)
```

Explanation:

- ord() returns ASCII value of each character
- map() applies ord() to every character
- sum() adds all ASCII values
- This total becomes the hash key

### Hash Table as a Class

We implement a simple hash table using:
- hash function
- buckets (lists)
- put() and get() methods

```python
class HashTable:
    def __init__(self, size):
        self.size = size
        self.table = {}

        for i in range(size):
            self.table[i] = []

    def hash(self, key):
        return key % self.size

    def get(self, key):
        return self.table[self.hash(key)]

    def put(self, key, value):
        bucket = self.table[self.hash(key)]
        if value not in bucket:
            bucket.append(value)

size = int(input("Enter hash table size: "))
ht = HashTable(size)

n = int(input("Enter number of values: "))

for i in range(n):
    value = input("Enter value: ")
    key = sum(map(ord, value)) 
    ht.put(key, value)

print("\nHash Table:")
for k, v in ht.table.items():
    print(k, ":", v)
```

### Roman Numeral Converter

Roman numerals use symbols:

I = 1  
V = 5  
X = 10  
L = 50  
C = 100  
D = 500  
M = 1000  

Rule:
- If smaller value comes before larger → subtract
- Otherwise → add

```python
def roman_to_int(s):
    roman = {
        'I': 1, 'V': 5, 'X': 10,
        'L': 50, 'C': 100,
        'D': 500, 'M': 1000
    }

    total = 0
    prev = 0

    for ch in reversed(s):
        value = roman[ch]

        if value < prev:
            total -= value
        else:
            total += value

        prev = value

    return total

s = input("Enter Roman numeral: ").upper()
print("Integer value:", roman_to_int(s))
```
