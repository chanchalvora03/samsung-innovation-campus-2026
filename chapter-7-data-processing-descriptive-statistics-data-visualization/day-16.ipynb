## Unit 35: Pandas Series

A **Series** is a one-dimensional labeled array in Pandas.

It is similar to a list, but with **index labels** for each element.

### Key Features:
- Stores single column of data
- Has index + values
- Can store different data types
- Supports fast operations

```python
import pandas as pd
data = [10, 20, 30, 40]
s = pd.Series(data)

print(s)
print(s[0])     # by position
```

```python
data = [10, 20, 30]
s = pd.Series(data, index=["a", "b", "c"])

print(s)

print(s["a"])   # by label
```

```python
s = pd.Series([1, 2, 3, 4])

print(s + 5)
print(s * 2)
```

```python
data = {"A": 90, "B": 85, "C": 88}
s = pd.Series(data)

print(s)

print(s.head())
print(s.tail())
print(s.mean())
print(s.max())
print(s.min())
```

```python
s = pd.Series([45, 67, 89, 34])

print(s[s > 50])
```

## Unit 37: Pandas DataFrame

A **DataFrame** is a 2D data structure (rows + columns), like a table.

It is made up of multiple Series.

### Features:
- Rows and columns
- Different data types
- Easy data manipulation
- Used in real-world datasets

```python
import pandas as pd

data = {
    "Name": ["Chanchal", "Bareeq", "Stacy"],
    "Age": [20, 22, 21],
    "Marks": [85, 90, 88]
}

df = pd.DataFrame(data)
print(df)
```

```python
print(df.head())     # first rows
print(df.tail())     # last rows
print(df.info())     # structure
print(df.describe()) # stats
```

```python
print(df["Name"])              # column
print(df.loc[0])              # row
print(df[["Name", "Marks"]])  # multiple columns
```

```python
print(df[df["Marks"] > 85])
```

```python
df["Grade"] = ["B", "A", "A"]
print(df)
```

```python
df_sorted = df.sort_values(by="Marks", ascending=False)
print(df_sorted)
```

```python
import numpy as np

df2 = pd.DataFrame({
    "A": [1, 2, np.nan],
    "B": [4, np.nan, 6]
})

print(df2.isnull())
print(df2.dropna())
print(df2.fillna(0))
```

```python
df1 = pd.DataFrame({"A": [1,2], "B":[3,4]})
df2 = pd.DataFrame({"A": [1,2], "C":[5,6]})

merged = pd.merge(df1, df2, on="A")
print(merged)
```

### Practise Problem

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df3 = pd.DataFrame({
    "Name": ["Alicia", "John", "Emma", "John"],
    "Age": [20, 22, np.nan, 22],
    "Marks": [85, 90, 88, 90],
    "City": ["Goa", "Delhi", "Mumbai", "Delhi"]
})

print("Initial DataFrame:\n", df3)

print("\nHead:\n", df3.head())
print("\nTail:\n", df3.tail())
print("\nInfo:")
print(df3.info())
print("\nDescribe:\n", df3.describe())

print("\nColumn (Name):\n", df3["Name"])
print("\nRow (loc[0]):\n", df3.loc[0])
print("\nRow (iloc[0]):\n", df3.iloc[0])
print("\nMultiple Columns:\n", df3[["Name", "Marks"]])

print("\nMarks > 85:\n", df3[df3["Marks"] > 85])

df3["Grade"] = ["B", "A", "A", "A"]
print("\nAfter Adding Grade:\n", df3)

del df3["Grade"]
print("\nAfter Deleting Grade:\n", df3)

print("\nMissing Values:\n", df3.isnull())
print("\nDrop NA:\n", df3.dropna())
df3_filled = df3.fillna(0)
print("\nFill NA with 0:\n", df3_filled)

print("\nRemove Duplicates:\n", df3.drop_duplicates())

print("\nSorted by Marks:\n", df3.sort_values(by="Marks", ascending=False))

print("\nReset Index:\n", df3.reset_index(drop=True))

df3.rename(columns={"Marks": "Score"}, inplace=True)
print("\nAfter Renaming:\n", df3)

df3.drop("Age", axis=1, inplace=True)
print("\nAfter Dropping Age:\n", df3)

print("\nGroup by City:\n", df3.groupby("City").sum(numeric_only=True))

extra = pd.DataFrame({
    "Name": ["Alicia", "John"],
    "Bonus": [5, 10]
})

merged = pd.merge(df3, extra, on="Name", how="left")
print("\nMerged Data:\n", merged)

df3.to_csv("output.csv", index=False)

df_loaded = pd.read_csv("output.csv")
print("\nLoaded CSV:\n", df_loaded.head())

# Bar Graph
df3.groupby("City")["Score"].mean().plot(kind="bar")
plt.title("Average Score by City")
plt.show()

# Line Graph
df3["Score"].plot(kind="line")
plt.title("Score Trend")
plt.show()

# Pie Chart
df3["City"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("City Distribution")
plt.show()
```

## Data Visualization using Matplotlib

Matplotlib is a Python library used to create graphs and charts.

### Types of graphs:
- Line Plot
- Bar Chart
- Pie Chart

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

plt.plot(x, y)
plt.title("Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
names = ["A", "B", "C"]
marks = [85, 90, 88]

plt.bar(names, marks)
plt.title("Marks Chart")
plt.show()
```

```python
labels = ["A", "B", "C"]
sizes = [30, 40, 30]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.show()
```
