```python
lst = [1,1,2,3,1]
print(lst.count(1)) # count() counts number of occurrences
```

```python
t = (1,2,5,4,3,2,1,4,7,8,9,9,3,7,3,9)

max_count = 0
result = None

for x in t:
    c = t.count(x)
    if c > max_count or (c == max_count and x > result):
        max_count = c
        result = x

print(result)
```

Found most frequent element using count and comparison logic  
Handled tie case by selecting larger value

### 2D List

```python
import random

seats = []

for i in range(3):
    row = []
    for j in range(6):
        row.append(random.randint(0,1))
    seats.append(row)

empty = 0

for row in seats:
    print(row)
    for seat in row:
        if seat == 0:
            empty += 1

print("Empty seats:", empty)
```

Simulated seating arrangement using 2D list  
Counted available seats using nested loops
