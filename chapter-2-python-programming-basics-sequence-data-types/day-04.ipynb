```
text="Samsung Innovation Campus"
print(text)
print(len(text))
print(text.upper())
print(text.lower())
print(text.replace("Campus","Program"))
```

## String Slicing

```
word="PYTHON"
print(word[0])
print(word[-1])
print(word[1:5])
print(word[::-1])
```

## Lists

```
fruits=["Apple","Banana","Mango"]
print(fruits)
fruits.append("Orange")
fruits.insert(1,"Grapes")
print(fruits)
fruits.remove("Banana")
print(fruits)
```

## Traversing a List

```
numbers=[10,20,30,40,50]
for n in numbers:
    print(n)
```

## Tuples

```
colors=("Red","Green","Blue")
print(colors)
print(colors[1])
print(len(colors))
```

## Practice

```
nums=[12,5,18,7,25]
print("Maximum =",max(nums))
print("Minimum =",min(nums))
print("Sum =",sum(nums))
```
