## for Loop

```
for i in range(1,6):
    print(i)
```

## while Loop

```
count=1
while count<=5:
    print(count)
    count+=1
```

## range() Examples

```
print(list(range(5)))
print(list(range(2,8)))
print(list(range(1,11,2)))
```

## break

```
for i in range(1,11):
    if i==6:
        break
    print(i)
```

## continue

```
for i in range(1,11):
    if i%2==0:
        continue
    print(i)
```

## pass

```
for i in range(3):
    pass
print("Loop completed")
```

## Nested Loops

```
for i in range(1,4):
    for j in range(1,4):
        print(f"({i},{j})", end=" ")
    print()
```

## Pattern Printing

```
rows=5
for i in range(1,rows+1):
    print("*"*i)
```

## Practice

```
n=int(input("Enter a number: "))
fact=1
for i in range(1,n+1):
    fact*=i
print("Factorial =",fact)
```
