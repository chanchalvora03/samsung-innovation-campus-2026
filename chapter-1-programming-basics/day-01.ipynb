## Variables

Variables are used to store data in memory.

```python
name = "Chanchal"
age = 20
college = "Christ University"

print(name)
print(age)
print(college)
```

## Data Types

```python
name = "Python"
version = 3.13
students = 35
is_fun = True

print(type(name))
print(type(version))
print(type(students))
print(type(is_fun))
```

## Arithmetic Operators

```python
a = 20
b = 10

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Exponent:", a ** b)
```
