## 1. Variables

```python
name="Chanchal"
age=20
cgpa=9.63
print("Name:",name)
print("Age:",age)
print("CGPA:",cgpa)
```

## 2. Variable Naming Rules
- Use letters, digits, underscores
- Cannot start with a digit
- Case sensitive

```python
student_name="Anish"
student_age=21
print(student_name)
print(student_age)
```

## 3. Data Types

```python
integer_value=25
float_value=10.75
string_value="Samsung"
boolean_value=True
for v in [integer_value,float_value,string_value,boolean_value]:
    print(type(v))
```

## 4. User Input

```python
name=input("Enter your name: ")
print("Welcome",name)
```

## 5. Type Casting

```python
num1=int(input("Enter first number: "))
num2=int(input("Enter second number: "))
print("Addition =",num1+num2)
```

## 6. Operators

```python
a,b=15,4
print(a+b,a-b,a*b,a/b,a//b,a%b,a**b)
```

## 7. Conditional Statements

```python
marks=int(input("Enter your marks: "))
if marks>=90:
    print("Grade A")
elif marks>=75:
    print("Grade B")
elif marks>=50:
    print("Grade C")
else:
    print("Fail")
```
